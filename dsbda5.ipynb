```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("heart.csv")
```

```python
#1. Data Cleaning
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Remove columns with more than 50% missing values
df.dropna(thresh=0.5*len(df), axis=1, inplace=True)
```

```python
# Check for duplicate data
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.info()
```

```python
# 2.Data Integration (first split dataset and then merge )
```

```python
#subset1
subset1= df[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs']].loc[0:]
subset1
```

```python
#Subset 2
subset2= df[['age', 'restecg', 'thalach', 'exang', 'oldpeak', 'ca']].loc[0:]
subset2
```

```python
merged_df = pd.merge(subset1, subset2, on='age')
```

```python
merged_df
```

```python
#Data Transformation(converting target column to binary values 0 & 1)
```

```python
df['target'] = df['target'].apply(lambda x: 1 if x>0 else 0)
```

```python
df.tail()
```

```python
#Error correcting
```

```python
df[df['age']<0] = df['age'].mean()
```

```python
#Data Modelling
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(['target'], axis=1)
y = df['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
